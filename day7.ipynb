```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Calculate Z-scores for numeric columns
z_scores = stats.zscore(df.select_dtypes(include='number'))

# Identify outliers (rows where any Z-score is greater than 3)
outliers = (z_scores > 3).any(axis=1)

# Count the number of outliers per column
outliers_count = outliers.sum()

# Create a bar chart to visualize the outliers
plt.figure(figsize=(10, 6))
plt.bar(['No Outliers', 'Outliers'], [len(df) - outliers_count, outliers_count])
plt.xlabel('Outlier Status')
plt.ylabel('Number of Rows')
plt.title('Number of Outliers')
plt.show()
```

```python
# Calculate Z-scores for numeric columns
z_scores = stats.zscore(df)

# Define a threshold for outlier detection (e.g., Z-score greater than 3)
threshold = 3

# Identify outliers (rows where any Z-score is greater than the threshold)
outliers = (z_scores > threshold).any(axis=1)

# Separate the DataFrame into two DataFrames: one with outliers and one without outliers
df_with_outliers = df[outliers]
df_without_outliers = df[~outliers]

# Display the DataFrame with outliers
print("DataFrame with Outliers:")
print(df_with_outliers)

# Display the DataFrame without outliers
print("\nDataFrame without Outliers:")
print(df_without_outliers)
```

```python
import pandas as pd

def remove_outliers(df, threshold=1.5):
    df_no_outliers = df.copy()  # Create a copy of the original DataFrame
    for col_name in df.columns:
        if pd.api.types.is_numeric_dtype(df[col_name]):
            q1 = df[col_name].quantile(0.25)
            q3 = df[col_name].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - threshold * iqr
            upper_bound = q3 + threshold * iqr
            df_no_outliers = df_no_outliers[(df_no_outliers[col_name] >= lower_bound) & (df_no_outliers[col_name] <= upper_bound)]
    return df_no_outliers
threshold_remove = 1.5

print("Before removing outliers:")
print(df.describe())

data_no_outliers = remove_outliers(df, threshold_remove)

print("\nAfter removing outliers:")
print(data_no_outliers.describe())
```
